III. Algorithm Designed
Ý tưởng:

Số Fibonacci thứ i được tính theo công thức

$F[1] = F[2] = 1$
$F[i] = F[i-1] + F[i-2]$
$\Rightarrow $Ta có ma trận hệ số:$ \begin{bmatrix}0 &amp; 1 \\ 1 &amp; 1 \end{bmatrix}$ và ma trận giá trị cơ sở: $ \begin{bmatrix} 1 \\  1 \end{bmatrix}$

$\Rightarrow  \begin{bmatrix}0 &amp; 1 \\ 1 &amp; 1 \end{bmatrix}^{k-1} \times  \begin{bmatrix} 1 \\  1 \end{bmatrix} = \begin{bmatrix} f[k] \\  f[k+1] \end{bmatrix} $

Không gian lưu trữ rất nhỏ không đáng kể

Độ phức tạp O($log(K)$)

$\Rightarrow$ Thỏa mản yêu cầu về độ phức tạp và không gian lưu trữ của bài toán

**Code**


```python
# Nhân ma trận
def mul(a,b):                           
	res = [[0]*len(b[0]) for u in range(len(a))]

	for i in range(len(a)) :
	 	for j in range(len(b[0])) :
	 		for k in range(len(a[0])) :
	 			res[i][j] = (res[i][j] + a[i][k] * b[k][j]) % 1000000007

	return res

def pow(a, n):                          
	res = a
	n -= 1
	c = a
	while n>0:
		if ((n&1) == 1):                    
			res = mul(res,c)
		c = mul(c,c)                        
		n = n>>1                            
	return res

n,k = map(int,input().split())

fibo   =  [[1],[1]]
matrix = [[0,1],[1,1]]

matrix = pow(matrix,k*2)                
fibo  = mul(matrix,fibo)                

res = (fibo[0][0] * n) % 1000000007    

# Xuất kết quả
print(res)
```
